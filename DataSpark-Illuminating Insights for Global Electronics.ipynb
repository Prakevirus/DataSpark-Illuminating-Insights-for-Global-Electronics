Handling Missing value with appropriate value

```python


# Fill missing 'State Code' with 'Unknown'
customer_analysis['State Code'] = customer_analysis['State Code'].fillna('Unknown')

# Convert 'Zip Code' to numeric and fill missing values with 0
customer_analysis['Zip Code'] = pd.to_numeric(customer_analysis['Zip Code'], errors='coerce')
customer_analysis['Zip Code'] = customer_analysis['Zip Code'].fillna(0)

# Ensure 'Delivery Date' and 'Order Date' are in datetime format
sales_analysis['Delivery Date'] = pd.to_datetime(sales_analysis['Delivery Date'], errors='coerce')
sales_analysis['Order Date'] = pd.to_datetime(sales_analysis['Order Date'], errors='coerce')

#convert USD into normal price
product_analysis["Unit Cost USD"] = product_analysis["Unit Cost USD"].str.replace("$", "").str.replace(",", "")
product_analysis["Unit Cost USD"] = product_analysis["Unit Cost USD"].astype(float)

product_analysis["Unit Price USD"] = product_analysis["Unit Price USD"].str.replace("$", "").str.replace(",", "")
product_analysis["Unit Price USD"] = product_analysis["Unit Price USD"].astype(float)

# Calculate the difference in days
day = (sales_analysis['Delivery Date'] - sales_analysis['Order Date']).dt.days

# Compute the mean of the differences in days
mean_day = day.mean()

# Fill missing 'Delivery Date' values with the mean difference
sales_analysis['Delivery Date'] = sales_analysis['Delivery Date'].fillna(sales_analysis['Order Date'] + pd.to_timedelta(mean_day, unit='D'))

# Convert 'Delivery Date' to date if needed
sales_analysis['Delivery Date'] = pd.to_datetime(sales_analysis['Delivery Date']).dt.date

# Fill missing values in 'Square Meters' with 0
store_analysis['Square Meters'] = store_analysis['Square Meters'].fillna(0)
print("Completed")
```
